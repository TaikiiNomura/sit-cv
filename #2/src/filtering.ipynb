```python
!uv add opencv-python
import numpy as np
from matplotlib import pyplot as plt 
import cv2
```

```python
img=cv2.imread('lenna.jpg')
plt.imshow(img)
```

```python
# フィルタ処理
print("平均化")
kernel=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)
plt.imshow(dst)
```

```python
dst=cv2.GaussianBlur(img,(3,3),-1)
plt.imshow(dst)
```

```python
img2=cv2.imread('lenna_color.png')
plt.imshow(img2)
```

```python
img2_=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
dst=cv2.medianBlur(img2_,3,-1)
plt.imshow(dst)
```

```python
img2__=cv2.GaussianBlur(img2_,(11,11),10)
dst=cv2.addWeighted(img2_,2,img2__,-1.0,0)
plt.imshow(dst)
```
